```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
import os

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = os.path.join("Resources", "menu_data.csv")
sales_filepath = os.path.join("Resources", "sales_data.csv")

# @TODO: Initialize list objects to hold our menu and sales data
menu = []
sales = []
menu_count=0
sales_count=0

# @TODO: Read in the menu data into the menu list
with open(menu_filepath, "r") as menufile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    menureader = csv.reader(menufile, delimiter=",")

    # Read the header row first 
    menu_header = next(menureader)
    # Print the header in suitable format 
    print('       | '.join(menu_header))
       
    # Read each row of data after the header
    for row in menureader:
        menu_count +=1
        menu.append(row)   
    print(menu_count)


# @TODO: Read in the sales data into the sales list
with open(sales_filepath, "r") as salesfile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    salesreader = csv.reader(salesfile, delimiter=",")

    # Read the header row first 
    sales_header = next(salesreader)
    # Print the header in suitable format 
    print('       | '.join(sales_header))

    
    # Read each row of data after the header
    for row in salesreader:
        sales_count +=1
        sales.append(row)
    print(sales_count) 


# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
report = {}

# Initialize a row counter variable
row_count = 0
new_item ={ "01-count": 0 , "02-revenue": 0 , "03-cogs": 0 , "04-profit": 0 }
menu_row=[]
sales_row=[]

# @TODO: Loop over every row in the sales list object
for sales_row in sales:

    # Line_Item_ID,Date,Credit_Card_Number,Quantity,Menu_Item
    # @TODO: Initialize sales data variables
    current_quantity =0
    current_revenue =0
    current_cost =0
    current_profit =0
    quantity= 0
    
    # @TODO:
    # If the item value not in the report, add it as a new entry with initialized metrics
    # Naming convention allows the keys to be ordered in logical fashion, count, revenue, cost, profit
    if sales_row[-1] not in report.keys():
        report[f"{sales_row[-1]}"] = new_item

    # @TODO: For every row in our sales data, loop over the menu records to determine a match
    for menu_row in menu:
           
                                                
        # Item,Category,Description,Price,Cost
        # @TODO: Initialize menu data variables
         if menu_row[0] == sales_row[-1]:
                price= int(menu_row[3])
                cost=int(menu_row[4])
                profit= price-cost
                quantity= int(sales_row[-2])


        # @TODO: Calculate profit of each item in the menu data
                current_quantity=int(report[sales_row[-1]]["01-count"])
                current_revenue=int(report[sales_row[-1]]["02-revenue"])
                current_cost=int(report[sales_row[-1]]["03-cogs"])
                current_profit=int(report[sales_row[-1]]["04-profit"])
                                
                


        # @TODO: If the item value in our sales data is equal to the any of the items in the menu, then begin tracking metrics for that item


            # @TODO: Print out matching menu data


            # @TODO: Cumulatively add up the metrics for each item key

                report[sales_row[-1]]["01-count"] = str(quantity + current_quantity)
                report[sales_row[-1]]["02-revenue"] = str((price * quantity) + current_revenue)
                report[sales_row[-1]]["03-cogs"] = str((cost * quantity)+ current_cost)
                report[sales_row[-1]]["04-profit"] = str((profit * quantity)+ current_profit)
        

                print(f" {row_count} : {sales_row[-1]} : {report} \n\n")


        # @TODO: Else, the sales item does not equal any fo the item in the menu data, therefore no match



    # @TODO: Increment the row counter by 1
    row_count += 1
    if row_count == 20:
        break
# @TODO: Print total number of records in sales data
print("-"*30)
print(f"Total Sales record: {row_count}")
print("----------- Report -----------")
for key, value in report.items():
    print(f"{key}  {value}")

# @TODO: Write out report to a text file (won't appear on the command line output)
with open('results.txt', 'w') as file:
    for key, value in report.items():
        file.write(f"{key}  {value}")
```
